```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/List_of_largest_companies_in_India"
page = requests.get(url)
```

```python
Soup = BeautifulSoup(page.text,'html')
Soup
```

```python
Soup.find_all(['h1','h2','h3'])
```

```python
paragraph = Soup.find_all('p')
print(paragraph)
```

```python
for i in paragraph:
    print(i.get_text().strip())
```

```python
Soup.find_all('table')[0]
```

```python
table = Soup.find_all('table')
print(table)
```

```python
table = Soup.find('table')  
headers = table.find_all('th')
print(headers)
```

```python
headers = [title.text.strip() for title in headers]
print(headers)
```

```python
df = pd.DataFrame(columns = headers)
df
```

```python
companies_data = table.find_all('tr')
```

```python
columns = ['Rank','Forbes 2000 rank','Name','Headquarters','Revenue(billions US$)','Profit(billions US$)','Assets(billions US$)','Value(billions US$)','Industry']  
df = pd.DataFrame(columns=columns)
all_data = []

for data in companies_data:
    DATA = data.find_all('td')
    Data_row_data = [d.text.strip() for d in DATA]
    if len(Data_row_data) == len(columns):
        all_data.append(Data_row_data)
    else:
        print(f"Skipping row due to column mismatch: {Data_row_data}")
df = pd.DataFrame(all_data, columns=columns)
```

```python
import pandas as pd
import seaborn as sns
sns.set(color_codes = True)
```

```python
file_path ='D:\Internship\Sharpended Mind\Project\Largest_companies.csv'
df = pd.read_csv(file_path)
print(df)
```

```python
df.head()
```

```python
df.info()
```

```python
industry_revenue = df.groupby('Industry')['Revenue(billions US$)'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Industry', y='Revenue(billions US$)', data=industry_revenue)
plt.xticks(rotation=90)
plt.title('Total Revenue by Industry (Bar Plot)')
plt.ylabel('Total Revenue (billions US$)')
plt.xlabel('Industry')
plt.tight_layout()
plt.show()
```

```python
top_companies = df.sort_values(by='Revenue(billions US$)', ascending=False).head(15)
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', 
          '#ffb3e6', '#ff6666', '#c2f0c2', '#c2e0f0', '#ffccf2']
plt.figure(figsize=(10, 10))
plt.pie(
    top_companies['Revenue(billions US$)'], 
    labels=top_companies['Name'], 
    autopct='%2.2f%%', 
    startangle=5, 
    colors=colors
)
plt.title('Revenue Distribution of Top 15 Companies')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['Revenue(billions US$)'], bins=30, kde=True, color='red')
plt.title('Histogram of Revenue Distribution (Top Companies)')
plt.xlabel('Revenue (Billions US$)')
plt.ylabel('Frequency')
plt.show()
```

```python
top_companies = df.sort_values(by='Revenue(billions US$)', ascending=False).head(15)
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Revenue(billions US$)', 
    y='Profit(billions US$)', 
    data=top_companies, 
    s=100,
    hue='Name',  
    palette='viridis'
)
plt.title('Scatter Plot of Revenue vs Profit for Top 15 Companies')
plt.xlabel('Revenue (Billions US$)')
plt.ylabel('Profit (Billions US$)')
plt.legend(title='Company Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.lineplot(x='Rank', y='Revenue(billions US$)', data=df, marker='o', label='Revenue',color='orange')
sns.lineplot(x='Rank', y='Profit(billions US$)', data=df, marker='o', label='Profit', color='green')
plt.title('Revenue and Profit vs Rank (Top Largest_Companies)')
plt.xlabel('Rank')
plt.ylabel('Billions US$')
plt.yticks(rotation=90)
plt.legend()
plt.show()
```

```python
df_top10 = df.head(10)
plt.figure(figsize=(10, 6))

plt.fill_between(df_top10['Name'], df_top10['Revenue(billions US$)'], label='Revenue', alpha=0.5)
plt.fill_between(df_top10['Name'], df_top10['Profit(billions US$)'], label='Profit', alpha=0.5)
plt.fill_between(df_top10['Name'], df_top10['Assets(billions US$)'], label='Assets', alpha=0.5)
plt.fill_between(df_top10['Name'], df_top10['Value(billions US$)'], label='Value', alpha=0.5)

plt.xlabel('Company Name')
plt.ylabel('Billions US$')
plt.title('Area Chart of Revenue, Profit, Assets, and Value for Top 10 Companies')
plt.xticks(rotation=90)
plt.yticks(rotation=90)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='Industry', y='Revenue(billions US$)', data=df)
plt.xticks(rotation=90)
plt.title('Violin Plot of Revenue Distribution Across Industries')
plt.ylabel('Revenue (billions US$)')
plt.xlabel('Industry')
plt.tight_layout()
plt.show()
```

```python
industry_revenue = df.groupby('Industry')['Revenue(billions US$)'].sum().reset_index()
industry_revenue.sort_values(by='Revenue(billions US$)', ascending=False, inplace=True)

industry_revenue['Cumulative Revenue'] = industry_revenue['Revenue(billions US$)'].cumsum()
industry_revenue['Cumulative Percentage'] = 100 * industry_revenue['Cumulative Revenue'] / industry_revenue['Revenue(billions US$)'].sum()

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.bar(industry_revenue['Industry'], industry_revenue['Revenue(billions US$)'], color='skyblue', label='Revenue')
ax1.set_ylabel('Revenue (billions US$)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_xticklabels(industry_revenue['Industry'], rotation=90, ha='right')

ax2 = ax1.twinx()
ax2.plot(industry_revenue['Industry'], industry_revenue['Cumulative Percentage'], color='orange', marker='o', label='Cumulative Percentage')
ax2.set_ylabel('Cumulative Percentage (%)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title('Pareto Chart of Revenue by Industry')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```
